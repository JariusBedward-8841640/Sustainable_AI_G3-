## ðŸ”· Section 2: Energy Prediction ML (Mostafa Allahmoradi)

### Key Contributions:
- **EnergyEstimator Class**: 326 lines of production code
- **Model Options**: Random Forest Regressor (100 trees), Linear Regression
- **Anomaly Detection**: Isolation Forest for flagging unusual patterns
- **Model Persistence**: Auto-save/load with joblib

````python
# =============================================================================
# 2.1 Energy Estimator Demonstration
# =============================================================================
from src.prediction.estimator import EnergyEstimator

print("=" * 70)
print("âš¡ ENERGY PREDICTION ENGINE")
print("=" * 70)

# Initialize the estimator (auto-loads trained model)
estimator = EnergyEstimator(model_type="LinearRegression")

print(f"\nðŸ¤– Model Type: {estimator.model_type}")
print(f"ðŸ“Š Features: {estimator.FEATURES}")
print(f"ðŸŽ¯ Target: {estimator.TARGET}")

# Test prediction
test_prompt = "Explain machine learning algorithms in detail with examples"
results = estimator.estimate(
    prompt_text=test_prompt,
    layers=12,
    training_hours=5.0,
    flops_str="1.5e18"
)

print(f"\nðŸ“ Test Prompt: '{test_prompt}'")
print(f"\nðŸ“Š Prediction Results:")
print(f"   âš¡ Energy: {results['energy_kwh']} kWh")
print(f"   ðŸŒ Carbon: {results['carbon_kg']} kgCO2")
print(f"   ðŸ“ Tokens: {results['token_count']}")
print(f"   ðŸ’¡ Suggestion: {results['suggestion']}")
````

## ðŸ”· Section 3: NLP & Prompt Optimizer (Oluwafemi Lawal)

### Key Contributions:
- **T5PromptOptimizer**: 1,148 lines - Fine-tuned T5-small transformer
- **Training Pipeline**: 195 examples, 5 epochs, ~25 min training time
- **Semantic Validation**: Sentence-Transformers (all-MiniLM-L6-v2)
- **Rule-Based Fallback**: Word replacements, filler removal

````python
# =============================================================================
# 3.1 Prompt Optimizer Demonstration
# =============================================================================
from src.nlp.simplifier import PromptSimplifier

print("=" * 70)
print("ðŸ§  NLP PROMPT OPTIMIZATION")
print("=" * 70)

# Initialize the simplifier
simplifier = PromptSimplifier(use_ml_model=True)

# Get model status
status = simplifier.get_model_status()
print(f"\nðŸ“Š Model Status:")
print(f"   T5 Available: {status.get('t5_available', False)}")
print(f"   Is Trained: {status.get('is_trained', False)}")

# Test optimization
test_prompts = [
    "Could you please help me understand how machine learning algorithms work in detail?",
    "I was wondering if you could explain the concept of neural networks step by step.",
    "Basically, I would really like to know about data preprocessing techniques."
]

print("\n" + "=" * 70)
print("ðŸ“ OPTIMIZATION EXAMPLES")
print("=" * 70)

for i, prompt in enumerate(test_prompts, 1):
    print(f"\nðŸ”¹ Example {i}:")
    analysis = simplifier.get_full_analysis(prompt)
    
    print(f"   Original ({analysis['original_tokens']} tokens):")
    print(f"   '{prompt[:60]}...'")
    print(f"\n   Optimized ({analysis['optimized_tokens']} tokens):")
    print(f"   '{analysis['optimized'][:60]}...'")
    print(f"\n   ðŸ“‰ Token Reduction: {analysis['token_reduction_pct']}%")
    print(f"   ðŸŽ¯ Semantic Similarity: {analysis['semantic_similarity']:.2f}")
````
